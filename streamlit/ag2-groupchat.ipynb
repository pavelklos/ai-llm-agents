# AG2 - GroupChat, GroupChatManager

## SETUP

```python
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

llm_config = {
    "api_type": "openai",
    "model": "gpt-4o-mini",
    "api_key": os.environ["OPENAI_API_KEY"]
}
```

## GroupChat, GroupChatManager

```python
# 1. Import necessary components
from autogen import ConversableAgent, GroupChat, GroupChatManager

# 2. Create your agents
teacher = ConversableAgent(
    name="teacher_agent",
    llm_config=llm_config,
    system_message="You are a teacher who decides lesson topics. Say DONE! when finished.",
    is_termination_msg=lambda x: "DONE!" in (x.get("content", "") or "").upper(),
)

planner = ConversableAgent(
    name="planner_agent",
    llm_config=llm_config,
    system_message="You create detailed lesson plans based on topics.",
)

reviewer = ConversableAgent(
    name="reviewer_agent",
    llm_config=llm_config,
    system_message="You review lesson plans against curriculum standards.",
)

# 3. Create GroupChat with agents
groupchat = GroupChat(
    agents=[teacher, planner, reviewer],
    speaker_selection_method="auto",  # Can also use "round_robin" or custom function
    messages=[],
)

# 4. Create GroupChatManager to manage the conversation
manager = GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config,
)
```

```python
# 5. Start the conversation
chat_result = teacher.initiate_chat(
    recipient=manager,
    message="Let's create a lesson about the solar system."
)
```

## ChatResult
